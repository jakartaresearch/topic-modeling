## Topic modelling using BERTopic

https://maartengr.github.io/BERTopic/tutorial/quickstart/quickstart.html

```python
import pandas as pd
import bertopic
from bertopic import BERTopic
from sentence_transformers import SentenceTransformer
```

```python
d_dataset = pd.read_json("../data/dataset/jan_sep_2020.json")
```

```python
d_dataset.head()
```

```python
d_dataset.shape
```

```python
d_dataset = d_dataset.loc[:9999, ]
```

```python
d_dataset.shape
```

```python
sentence_model = SentenceTransformer('stsb-xlm-r-multilingual')
```

```python
topic_model = BERTopic(language='indonesian', n_gram_range=(2, 3), min_topic_size=10,
                       calculate_probabilities=True, nr_topics="auto",
                       embedding_model=sentence_model)
```

```python
topics, probs = topic_model.fit_transform(d_dataset.title)
```

```python
topic_model.visualize_topics()
```

```python
topic_model.visualize_distribution(probs[0])
```
